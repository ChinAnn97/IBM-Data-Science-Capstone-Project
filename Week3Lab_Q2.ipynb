**After importing beautiful soup library, create an url to send a request to retrieve html file. Then, parse the html file.**

```python
# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, "html.parser")
tag = soup('table')
```

**Get all the headings.**

```python
table = tag[0]
table_data = table.tbody.find_all("tr")
```

```python
table_data[0]
```

```python
headings = []
for th in table_data[0].find_all("th"):
    headings.append(th.text.replace('\n', ' ').strip(" "))
print(headings)
```

**Iterate thorough the "tr" and "td" tags to retireve and store the information on postal code, borough, and neighborhoood.**

```python
i = 0
Postal_code = []
Borough = []
Neighbourhood = []
for tr in table_data[1:]:
    for td in tr.find_all("td"):
        if i == 0:
            Postal_code.append(td.text.replace('\n', ' ').strip())
        elif i== 1:
            Borough.append(td.text.replace('\n', ' ').strip())
        else:
            Neighbourhood.append(td.text.replace('\n', ' ').strip())
        i = i + 1
    i = 0
for code, borough, neighbourhood in zip(Postal_code,Borough,Neighbourhood):
    print("{}, {}, {} \n". format(code, borough, neighbourhood))
```

**Create a data frame using headings, postal code, borough, and neighbpurhood. Rename the columns, filter "borough" columns**

```python
import pandas as pd
df = pd.DataFrame({headings[0] : Postal_code, headings[1] : Borough, headings[2] : Neighbourhood})
df
```

```python
df.rename(columns = {"Postal Code" : "PostalCode", "Neighbourhood" : "Neighborhood"}, inplace = True)
df
```

```python
df_new = df[df["Borough"] != "Not assigned"]
df_new
```

```python
df_new.reset_index(inplace = True)
df_new
```

```python
df_new.drop(["index"], axis = 1, inplace =True)
df_new
```
